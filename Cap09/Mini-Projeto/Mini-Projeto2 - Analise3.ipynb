```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
```

## Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio

```python
# Crie um Barplot com o Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio
g, ax = plt.subplots(figsize=(8, 6))
colors = ['#E68191', '#53A144','#589BD5']
sns.set_style('whitegrid')
sns.barplot(x='fuelType', y='price', hue='gearbox', palette=colors, data=df)
ax.set_title('Preço médio do veículo por tipo de combustível e tipo de caixa de câmbio', fontdict={'size': 12})
ax.set_xlabel('Tipo de Combustível', fontdict={'size': 14})
ax.set_ylabel('Preço Médio', fontdict={'size': 14})
plt.show()
```

```python
# Salvando o plot
g.savefig("plots/Analise3/fueltype-vehicleType-price.png")
```

## Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio

```python
# Crie um Barplot com a Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio
g, ax = plt.subplots(figsize=(8, 6))
colors = ['#4E76AC', '#DCA81D', '#A4A296']
sns.set_style('whitegrid')
sns.barplot(x='vehicleType', y='powerPS', hue='gearbox', palette=colors, data=df)
ax.set_title('Potência média de um veículo por tipo de veículo e tipo de caixa de câmbio', fontdict={'size': 12})
ax.set_xlabel('Tipo de Veículo', fontdict={'size': 14})
ax.set_ylabel('Potência Média', fontdict={'size': 14})
plt.show()
```

```python
# Salvando o plot
g.savefig("plots/Analise3/vehicletype-fueltype-power.png")
```
